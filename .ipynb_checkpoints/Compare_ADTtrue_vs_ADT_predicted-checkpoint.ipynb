```python
from sklearnex import patch_sklearn
# Reimplementations of sklearn functions with optimization for speed
patch_sklearn(global_patch=True)
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import seaborn as sns; sns.set()
import pandas as pd
from matplotlib.collections import QuadMesh
from matplotlib.text import Text
import random
```

```python
gex_normed = pd.read_csv('Documents/GitRepos_local/scLinear_R/3TModel_test_gex_normed.txt',sep = '\t')
annot_seurat = pd.read_csv('Documents/GitRepos_local/scLinear_R/3TModel_test_metadata_celltype.txt',sep = '\t')
annot_seurat['datset'] = annot_seurat.index.to_series().str.extract("^([^_]*)_[ACTG]")
measured_adt_clr = pd.read_csv('Documents/GitRepos_local/scLinear_R/3TModel_test_true_ADT.txt',sep = '\t')
predicted_adt_clr = pd.read_csv('Documents/GitRepos_local/scLinear_R/3TModel_test_predicted_ADT.txt',sep = '\t')
```

```python
gex_normed = gex_normed.transpose()
annot = pd.Series(annot_seurat.cell_type)
train_size = 5000
min_sample_per_cell = annot.groupby(annot).apply(lambda x: x.sample(min(int(0.8 *x.size),20), random_state=1))
train_indices = set(min_sample_per_cell.index.get_level_values(1))
rest = set(annot.index).difference(train_indices)
train_indices = train_indices.union(set(random.sample(list(rest),train_size-len(train_indices))))
test_indices = set(annot.index).difference(train_indices)
train_indices = list(train_indices)
test_indices = list(test_indices)
```

```python
svc_adt_true = svm.SVC(cache_size = 4000)
svc_adt_true.fit(measured_adt_clr.loc[train_indices,:],annot[train_indices])

svc_adt_predi = svm.SVC(cache_size = 4000)
svc_adt_predi.fit(predicted_adt_clr.loc[train_indices,:],annot[train_indices])
```

```python
predictions_from_measured_adt = svc_adt_true.predict(measured_adt_clr.loc[test_indices,:])
correct = sum(predictions_from_measured_adt == annot[test_indices])
wrong = sum(predictions_from_measured_adt != annot[test_indices])
accuracy = correct/(wrong+correct)
print('{acc:.2f}% of labels in testset correctly assigned based on measured ADT'.format(acc = 100*(accuracy)))
```

```python
predictions_from_predicted_adt = svc_adt_predi.predict(predicted_adt_clr.loc[test_indices,:])
correct = sum(predictions_from_predicted_adt == annot[test_indices])
wrong = sum(predictions_from_predicted_adt != annot[test_indices])
accuracy = correct/(wrong+correct)
print('{acc:.2f}% of labels in testset correctly assigned based on predicted ADT'.format(acc = 100*(accuracy)))
```

```python
import re
gex_normed.index = gex_normed.index.to_series().str.replace(r'\.(\d+)$', r'-\1', regex=True)
import time
svc_full_gene= svm.SVC(cache_size = 4000)
stime = time.time()
svc_full_gene.fit(gex_normed.loc[train_indices,:],annot[train_indices])
etime = time.time()
print(etime-stime)
```

```python
test_predictions_gene_full = svc_full_gene.predict(gex_normed.loc[test_indices,:])
correct = sum(test_predictions_gene_full == annot[test_indices])
wrong = sum(test_predictions_gene_full != annot[test_indices])
accuracy = correct/(wrong+correct)
print('{acc:.2f}% of labels in testset correctly assigned based on full gene expression'.format(acc = 100*(accuracy)))
```

```python
def plot_crosstab(annot1,annot2,title):
    ctab = pd.crosstab(annot1,annot2,margins = True)
    ctab.dtype = int
    ctab = ctab.reindex(ctab.columns).fillna(0).astype(int)
    # fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    fig = plt.figure(figsize = (20,20))
    
    plt.clf()
    ax = fig.add_subplot(111)
    ax.set_aspect(1)
    res = sns.heatmap(ctab, annot=True,vmin= 0,vmax = 20,annot_kws={"fontsize":16},fmt='d',cbar = False)
    quadmesh = ax.findobj(QuadMesh)[0]
    facecolors = quadmesh.get_facecolors()
    facecolors = facecolors.reshape((ctab.shape[0],ctab.shape[1],4))
    facecolors[:,ctab.shape[1]-1,:] = np.ones((ctab.shape[1],4))
    facecolors[ctab.shape[0]-1,:,:] = np.ones((ctab.shape[0],4))

    # plt.title('How to plot a coningency table with python \n using matplotlib and seaborn ?',fontsize=12)
    ax.pcolormesh(facecolors)
    for i in ax.findobj(Text):
        i.set_color('black')
    plt.xticks(fontsize = 20,rotation = 90)
    plt.yticks(fontsize = 20, rotation = 0)
    plt.xlabel('Annotation')
    plt.ylabel('Prediction')
    # plt.savefig(title+".png", bbox_inches='tight', dpi=100)
    
    plt.show()
```

```python
plot_crosstab(annot_seurat_testcells['annot_from_measured_adt'],annot_seurat_testcells['cell_type'],'Seurat vs. based on measured ADT')
```

```python
plot_crosstab(annot_seurat_testcells['annot_from_predicted_adt'],annot_seurat_testcells['cell_type'],'Seurat vs. based on measured ADT')
```

```python
plot_crosstab(annot_seurat_testcells['reannot_from_rna'],annot_seurat_testcells['cell_type'],'Seurat vs. reannot based on RNA')
```
